```python
# creating another variable for updated data
updated_cars = cars
```

```python
# removing features which does not realte with price
updated_cars = updated_cars.drop(['car_ID','symboling','carheight','stroke','compressionratio','peakrpm','highwaympg'],axis=1)
```

```python
# final updated data
updated_cars
```

```python
# new correlatio diagram 
corr = updated_cars.corr()
plt.subplots(figsize=(10,10))
sns.heatmap(corr,annot=True)
```

```python
# splitting data into features and target values
X = updated_cars.drop(['price'],axis=1)
Y = updated_cars['price']
```

```python
#  converting a categorical variable into dummy/indicator variables
X = pd.get_dummies(data=X, drop_first=True)
```

```python
# importing train and test split funciton 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.1)
```

```python
# importing Linear Regression Model
from sklearn.linear_model import LinearRegression

# creating instance for logistic regression model
lr= LinearRegression()

# training the model on train dataset
lr.fit(X_train,Y_train)
```

```python
# getting predictions from the model
prediction = lr.predict(X_test)
```

```python
# importing and checking r2_score
from sklearn.metrics import r2_score
r2 = r2_score(prediction,Y_test)
r2
```
